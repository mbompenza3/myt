# Теперь рассмотрим статистику  Count,Sum в различных разрезах. По сути эти данные показывают информацию об объеме поступления денег

## Применяем метод groupby для группирования данных по Level и describe для вывода статистик. Видим, что в разрезе заданной таблицы, на 11 уровне   наибольшие Count, Sum  в среднем , максимальное  по одной строке, также по процентилю 50%.   

```python
columns_to_show = ['Count', 'Sum']

df.groupby(['Level'])[columns_to_show].describe(percentiles=[])
```

# Применяем метод groupby для группирования данных по магазинам и вывод статистик. Наибольшее Count, Sum в среднем, максимум в отдельной строке и 50% процентиль оказалось в Sale3. 

```python

df.groupby(['From'])[columns_to_show].describe(percentiles=[])
```

# Теперь выводим общую сумму  Count и Sum в разрезе уровней. Наибольшее число покупок 99436 и сумма покупок 725307 у 11 уровня, наименьшее у 1.

```python
df.groupby(['Level'])[columns_to_show].sum()
```

# Теперь выводим общую сумму  Count и Sum в разрезе 2 аналитик, в groupby указываем 2 поля: Level в паре с Type, Kach, VIP поочередно. Так можем видеть как каждый уровень тратит деньги по типу предмета, по качеству предмета, по наличию подписки

```python
df.groupby(['Level','Type'])[columns_to_show].sum()
```

```python
pd.set_option('display.max_rows', 100)
df.groupby(['Level','Kach'])[columns_to_show].sum()
```

```python
df.groupby(['Level','VIP'])[columns_to_show].sum()
```

# Применяя функцию nlargest можно узнать указанное кол-во отсортированных по убыванию значений по какому-либо полю. Например наибольшая сумма получена по 11 уровню по предметам с качеством 6. Вообще видно, что 4 первые записи (11 уровень, 6,9,7,8 качество) по сумме намного превосходят остальные пары Level, Kach  

```python
df.groupby(['Level','Kach'])[columns_to_show].sum().nlargest(10,['Sum'])       
```

# наибольшое кол-во также получено по 11 уровню по предметам с качеством 6.  4 первые записи (11 уровень, 6,9,7,8 качество) по кол-ву намного превосходят остальные пары Level, Kach  

```python
df.groupby(['Level','Kach'])[columns_to_show].sum().nlargest(10,['Count'])     
```

# Наибольшая сумма получена по 11 уровню по предметам с типом 2. Аналогично используя groupby при желании можно посмотреть лучшие продажи  по любым  парам, тройкам и т.д. аналитик 

```python
df.groupby(['Level','Type'])[columns_to_show].sum().nlargest(10,['Sum'])  
```

# Можно использовать функцию crosstab, для вывода кросстаблицы по паре аналитик. По паре Level, Kach видим корреляцию, чем выше уровень тем лучше качество покупок

```python
pd.crosstab(df['Level'], df['Kach'])
```

# По паре Level, Type видим что маленькие уровни в основном покупали тип 2, средние - тип 3, большие - тип 1. Самый высший 11 - тип 1.

```python
pd.crosstab(df['Level'], df['Type'])
```

# По Level, VIP : малые уровни примерно равнозначно, после 4 уровня видно что существенно больше покупок у имеющих подписку.

```python
pd.crosstab(df['Level'], df['VIP'])
```
